```python
# 匯入所需套件
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 設定中文顯示（避免圖表中文亂碼）
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

print("套件載入成功！")
```

```python
# 下載台積電近一年股價數據
# 台股代碼要加 .TW
tsmc = yf.download('2330.TW', start='2024-01-01', end='2025-02-09')

# 顯示前5筆數據
print("台積電股價數據：")
tsmc.head()
```

```python
# 繪製台積電收盤價走勢圖
plt.figure(figsize=(14, 7))
plt.plot(tsmc.index, tsmc['Close'], linewidth=2, color='#2E86AB')
plt.title('台積電(2330) 股價走勢圖', fontsize=18, fontweight='bold')
plt.xlabel('日期', fontsize=14)
plt.ylabel('收盤價 (TWD)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 下載聯發科股價
mtk = yf.download('2454.TW', start='2024-01-01', end='2025-02-09')

# 比較兩檔股票
plt.figure(figsize=(14, 7))
plt.plot(tsmc.index, tsmc['Close'], label='台積電(2330)', linewidth=2)
plt.plot(mtk.index, mtk['Close'], label='聯發科(2454)', linewidth=2)
plt.title('台積電 vs 聯發科 股價比較', fontsize=18, fontweight='bold')
plt.xlabel('日期', fontsize=14)
plt.ylabel('收盤價 (TWD)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
